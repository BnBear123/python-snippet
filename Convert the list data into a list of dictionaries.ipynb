```python
import pandas as pd
```

```python
df = pd.read_excel('charaterfair_email_interantional.xlsx')
```

```python
df.shape
```

```python
df.head(16)
```

```python
data = df['참가기업'].tolist()
```

```python
# Convert the list data into a list of dictionaries
records = []
current_record = {}

for i in range(0, len(data), 2):
    key, value = data[i], data[i+1]
    if key == "국가" and current_record:
        records.append(current_record)
        current_record = {}
    current_record[key] = value

records.append(current_record)
```

```python
# Creating the dataframe with the desired columns
df_final = pd.DataFrame(records, columns=["국가", "업체명(영문)", "대표자명(영문)", "대표 연락처", "대표 E-Mail"])
```

```python
df_final.to_csv('charaterfair_email(interantional).csv', index=False, encoding="utf-8-sig")
```
